```python
import pandas as pd
from prophet import Prophet
from prophet.plot import plot_plotly
```

```python
data = pd.read_csv("Data/DailyDelhiClimateTrain.csv")
```

```python
data
```

# Select and rename the date and temp column for prophet model

```python
selected_df = data[["date", "meantemp"]]
selected_df = selected_df.rename(columns={"date": "ds", "meantemp": "y"})
```

```python
selected_df
```

```python
last = selected_df[-20:]
first = selected_df[: -20]
```

```python
first
```

```python
last
```

```python
fbp = Prophet(yearly_seasonality=True)
```

```python
fbp.fit(first)
```

```python
future = fbp.make_future_dataframe(periods=365)
forecast = fbp.predict(future)
```

```python
forecast
```

```python
plot_plotly(fbp, forecast)
```
