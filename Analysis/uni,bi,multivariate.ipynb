```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/iris-data.csv")
```

```python
data
```

# Univarite Analysis

```python
data["class"].value_counts()
```

```python
data_setosa = data.loc[data["class"] == "Iris-setosa"]
data_versicolor = data.loc[data["class"] == "Iris-versicolor"]
data_virginica = data.loc[data["class"] == "Iris-virginica"]
```

```python
plt.plot(data_setosa["sepal length"], np.zeros_like(data_setosa["sepal length"]), 'o')
plt.plot(data_versicolor["sepal length"], np.zeros_like(data_versicolor["sepal length"]), 'o')
plt.plot(data_virginica["sepal length"], np.zeros_like(data_virginica["sepal length"]), 'o')
plt.show()
```

# Bivariate Analysis

```python
sns.FacetGrid(data, hue="class").map(plt.scatter, "petal length", "sepal length").add_legend()
plt.show()
```

```python
sns.FacetGrid(data, hue="class").map(plt.scatter, "petal width", "sepal length").add_legend()
plt.show()
```

```python
sns.FacetGrid(data, hue="class").map(plt.scatter, "sepal width", "sepal length").add_legend()
plt.show()
```

# Multivariate analysis

```python
sns.pairplot(data, hue="class", size=5)
```
