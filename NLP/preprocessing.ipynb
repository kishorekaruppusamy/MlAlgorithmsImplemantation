```python
df_text = full_df[["text"]]
```

```python
df_text["text"] = df_text["text"].astype(str)
```

```python
df_text["text"]
```

```python
df_text["text_lower"] = df_text["text"].str.lower()
df_text.head()
```

```python
PUNCTUATION_TO_REMOVE = string.punctuation

def remove_punctuation(text):
    return text.translate(str.maketrans('', '', PUNCTUATION_TO_REMOVE))

df_text["punc_removed"] = df_text["text_lower"].apply(lambda text: remove_punctuation(text))
df_text.head()
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
", ".join(stopwords.words("english"))
```

```python
STOPWORDS = set(stopwords.words("english"))

def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

df_text["stopword_removed"] = df_text["punc_removed"].apply(lambda text: remove_stopwords(text))
df_text.head()
```

```python
from nltk.stem.porter import PorterStemmer

stemmer = PorterStemmer()

def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])

df_text["stemming"] = df_text["stopword_removed"].apply(lambda text: stem_words(text))
df_text.head()
```

```python
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
def lemmatize_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])

df_text["lemmatization"] = df_text["stopword_removed"].apply(lambda text: lemmatize_words(text))
df_text.head()
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
from nltk.corpus import wordnet
wordNet_map = {"N": wordnet.NOUN, "V": wordnet.VERB, "J": wordnet.ADJ, "R": wordnet.ADV}
def lemmatize_words_with_parts_of_speech(text):
    pos_tagged_text = nltk.pos_tag(text.split())
    return " ".join([lemmatizer.lemmatize(word, wordNet_map.get(pos[0], wordnet.NOUN)) for word, pos in pos_tagged_text])

df_text["lemmatization_with_pos"] = df_text["lemmatization"].apply(lambda text: lemmatize_words_with_parts_of_speech(text))
df_text.head()
```
